```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("IMDB_processed_data.csv")

# --- 1. Cek Missing Value ---
print(" Jumlah Missing Value Tiap Kolom:")
print(df.isnull().sum())
print("\n")

# --- Handling Missing Value ---
# Contoh: isi nilai kosong dengan mean (untuk numerik)
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

# Untuk kolom kategorikal, isi dengan mode (nilai terbanyak)
object_cols = df.select_dtypes(include='object').columns
for col in object_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print(" Missing value sudah ditangani.")
print("\n")


# --- 2. Cek dan Tangani Duplikat ---
print(" Jumlah duplikat:", df.duplicated().sum())
df = df.drop_duplicates()
print(" Duplikat sudah dihapus.\n")


# --- 3. Handling Outlier ---
# Gunakan metode IQR (Interquartile Range)
Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

# Tentukan batas bawah dan atas
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Tandai dan buang outlier
outlier_mask = ((df[numeric_cols] < lower_bound) | (df[numeric_cols] > upper_bound))
outlier_rows = outlier_mask.any(axis=1)
print(f" Jumlah baris dengan outlier: {outlier_rows.sum()}")

df_clean = df[~outlier_rows]
print(f" Dataset bersih: {df_clean.shape[0]} baris, {df_clean.shape[1]} kolom\n")


# --- 4. Visualisasi Outlier (Boxplot Sebelum & Sesudah) ---
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.boxplot(data=df[numeric_cols])
plt.title("Sebelum Handling Outlier")

plt.subplot(1,2,2)
sns.boxplot(data=df_clean[numeric_cols])
plt.title("Sesudah Handling Outlier")

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd

# Ambil kolom numerik
numeric_cols = df_clean.select_dtypes(include=['int64', 'float64']).columns

# --- 1 Normalisasi (Min-Max Scaling) ---
minmax_scaler = MinMaxScaler()
df_normalized = df_clean.copy()
df_normalized[numeric_cols] = minmax_scaler.fit_transform(df_clean[numeric_cols])

print(" Normalisasi (Min-Max Scaling) selesai.")
print(df_normalized[numeric_cols].head())

# --- 2 Standarisasi (Z-Score Scaling) ---
standard_scaler = StandardScaler()
df_standardized = df_clean.copy()
df_standardized[numeric_cols] = standard_scaler.fit_transform(df_clean[numeric_cols])

print("\n Standarisasi (Z-Score Scaling) selesai.")
print(df_standardized[numeric_cols].head())
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Coba ubah kolom Runtime jadi numerik (hapus teks 'min' kalau ada)
df['Runtime'] = df['Runtime'].replace(' min', '', regex=True)  # hapus kata 'min' kalau ada
df['Runtime'] = pd.to_numeric(df['Runtime'], errors='coerce')  # ubah ke angka

# Cek ulang tipe datanya
print(df.dtypes)

# Sekarang pilih ulang kolom kategorikal
obj_cols = df.select_dtypes(include='object').columns
print("\nKolom kategorikal sesungguhnya:", obj_cols.tolist())

# Lakukan label encoding hanya pada kolom kategorikal yang benar
label_encoders = {}
df_encoded = df.copy()

for col in obj_cols:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le

print("\n Label Encoding selesai.")
print(df_encoded[obj_cols].head())
```
